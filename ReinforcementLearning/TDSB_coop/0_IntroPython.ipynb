Data Structures

```
# lists and arrays 
my_list = [1,2,3,4]
my_array = np.array(my_list)
print(f'examples of arrays: {my_list}, {my_array}')

# tuples
tup1 = (1,2)
tup2 = tuple(my_list)
print(f'examples of tuples: {tup1}, {tup2}')

# dictionaries
# keys must be strings or other 'immutable' types i.e. tuple
# entries can be anything
my_dict = {'key1':10, 'key2':'apple'}
my_dict['key3'] = (10,200)
my_dict[(5,6)] = my_list

print(f'an example dict: {my_dict}')
```
