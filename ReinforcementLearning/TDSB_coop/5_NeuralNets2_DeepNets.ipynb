# Neural networks are chained perceptrons

image.png

Neural networks are universal function approximators. This means that by following gradient descent learning rules, neural networks can adjust their weights so that they can produce a desired output given some input. This simple notion has extremely powerful implications. You can train a neural network to classify handwritten digits (MNIST dataset), to identify the likelihood a photo contains a car or a dog (CIFAR10/imagenet datasets), or even to navigate an environment (as we previously did with the Q-learners in the tabular case). In this tutorial, we will work through image classification to become more familiar with the basics of neural networks. In subsequent tutorials, we will use neural networks to learn good behaviour in a reinforcement learning task. 

A simple neural network has an input layer, a hidden layer, and an output layer. A deep neural network has multiple hidden layers. However, the basic principle is very similar. Any layer gets inputs from the layer before it. The activity of that layer is the activity of the layer before, multiplied by the weights between the two layers, and scaled by some activation function. This is very similar in practice to the perceptrons we looked at in the previous tutorial, except that we are looking at multiple units at once. 

```
import numpy as np
```

```
# as before with the perceptron we will make a random input vector and a random set of weights
num_inputs = 10         # number of units in the input layer
num_hidden_units = 20   # number of units in the hidden layer
num_outputs = 4         # number of units in the output layer

# weights is an array where rows correspond to the cells the connections are coming FROM
# and columns correspond to the cells the connections are going TO
# weights[i,j] is the weight from the ith cell in the input to the jth cell in the output
weights_1 = np.random.standard_normal((num_inputs,num_hidden_units))
weights_2 = np.random.standard_normal((num_hidden_units, num_outputs))
```

Above we have given the necessary information to generate the simple neural network pictured on the left. Using some random numbers as input, we can get a hidden layer activity and output activity in the following cell

```

inputs = np.random.random(num_inputs)
hidden_layer_activities = np.matmul(inputs,weights_1)
outputs = np.matmul(hidden_layer_activities,weights_2)

print(f' Input vector {inputs.shape}:\n {inputs} \n\n Hidden activity vector {hidden_layer_activities.shape}:\n {hidden_layer_activities} \n\n Output {outputs.shape}:\n {outputs}')
```
